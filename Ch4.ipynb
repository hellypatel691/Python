# Multiple Linear Regression

```python
import pandas as pd
# step1 read csv file
df=pd.read_csv('car data.csv')
df.head()
```

```python
# step2 check null value
df.isna().sum()
```

```python
# step3 data preprocessing
df['Age']=2026-df['Year']
df.head()
```

```python
df.drop(columns='Year',inplace=True)
df.head()
```

```python
df.drop(columns='Car_Name',inplace=True)
df.head()
```

```python
df=pd.get_dummies(data=df,drop_first=True)
df.head()
```

```python
# step4 define x and y
x2=df[['Present_Price', 'Kms_Driven', 'Owner', 'Age',
       'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Seller_Type_Individual',
       'Transmission_Manual']]
y2=df['Selling_Price']
```

```python
df.columns
```

```python
# step5 split data into train and test
from sklearn.model_selection import train_test_split
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.2,random_state=42)
print(x2_train.shape)
print(x2_test.shape)
print(y2_train.shape)
print(y2_test.shape)
```

```python
# step6 create model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x2_train,y2_train)
```

```python
# step7 find m and c
m=lr.coef_
m
```

```python
c=lr.intercept_
c
```

```python
df.columns
```

```python
# Enter data according to the above format of column
i20_price=lr.predict([[9,45000,0,8,0,1,1,1]])              
```

```python
maserati_price=lr.predict([[9.54,43000,0,13,1,0,0,1]])
print(maserati_price)
```

```python
y2_pred=lr.predict(x2_test)
y2_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print("MSE: ",mean_squared_error(y2_test,y2_pred))
print("MAE: ",mean_absolute_error(y2_test,y2_pred))
print("R2_Score: ",mean_squared_error(y2_test,y2_pred))
```
